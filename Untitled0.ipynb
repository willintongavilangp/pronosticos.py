```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
from sklearn.metrics import mean_squared_error ,r2_score
```

```python
# ingreso de datos de numero de mes en x, montos de ventas en Y e mes a pronosticar en Z:
x = [1,2,3,4,5,6]
y = [7000, 9000, 5000, 11000, 10000, 13000]
z = [7]

# hay que convertir las listas en matrices:

x = np.array(x)
y = np.array(y)
z = np.array(z)

# para graficar los datos de entrada:

plt.scatter(x,y,label='data',color='red')
plt.title('Distribucion entre meses y ventas');
```

```python
# en esta etapa vamos a entrenar el modelo para hacer la regresion lineal

regresion_lineal = linear_model.LinearRegression()
regresion_lineal.fit(x.reshape(-1,1),y)

# ahor hay que imprimir los parametro que estimo la regresion

print('\nparametros del modelo de regresion')
print('b (pendiente) = ' + str(regresion_lineal.coef_) +' , a (punto de corte) = ' + str(regresion_lineal.intercept_))
```

```python
pronostico = regresion_lineal.predict(z.reshape(-1,1))

print('\npronosticos')
for i in range(len(z)):
  print('pronostico para el periodo{0} = {1} '.format(z[i], pronostico[i]))
```

```python
# para hacer la evaluacion de la regresion

pronostico_entrenamiento = regresion_lineal.predict(x.reshape(-1,1))
mse = mean_squared_error(y_true= y, y_pred= pronostico_entrenamiento)
rmse = np.sqrt(mse)
print('\nevaluacion de calidad de la regresion')
print('Error Cuadratico Medio (MSE) = ' + str(rmse))

r2 = regresion_lineal.score(x.reshape(-1,1), y)
print('coeficiente de determinacion R2 ='+ str(r2))
```

```python
plt.plot(x,pronostico_entrenamiento, label='data' , color='blue')
plt.xlabel('Meses')
plt.ylabel('ventas')
```
